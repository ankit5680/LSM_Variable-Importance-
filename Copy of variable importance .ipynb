```python
 data= pd.read_csv('check1.csv')
print(data.head())
```

```python
axis = data.columns.get_loc('CID')

print("Axis of 'Movement column:", axis)
```

```python
X = data.drop('CID', axis=1)  # Assuming 'landslide' is the column name for the target variable
y = data['CID']
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=100)
```

```python
rf.fit(X, y)
```

```python
importances = rf.feature_importances_
```

```python
# Sort the feature importances in descending order
sorted_indices = importances.argsort()[::-1]

# Print the variable importance ranking
print("Variable Importance Ranking:")
for i, index in enumerate(sorted_indices):
    print(f"{i+1}. {X.columns[index]}: {importances[index]}")
```
